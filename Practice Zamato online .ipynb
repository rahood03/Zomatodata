```python
# Groupinig all columns with rating. 

final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
# Groupinig all columns with rating, and doing the reset functions.
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
# Groupinig all columns with rating, and doing the reset functions, and renaming the 0 column to rating count

ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
#Arranging rating in decending order.(Aggregate rating)

ratings_decending=ratings.sort_values('Aggregate rating',ascending=False)
ratings_decending
```

```python
#Arranging rating in decending order.(Aggregate rating)

rating_ascending=ratings.sort_values('Aggregate rating', ascending=True)
rating_ascending
```

##Observation

1 Between Aggregrate rating 4.5-4.9, it indicate Excellent. 
2 Between Aggregrate rating 4.0-4.4, it indicate Very Good
3 Between Aggregrate rating 3.5-3.9, it indicate Good
4 Between Aggregrate rating 2.5- 3.4 , it indicate Average
5 Between Aggregrate rating 1.8- 2.4, it indicate poor
6 0.0 didnt rate.
